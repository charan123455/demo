# IMPUTING THE MISSING VALUES

# REPLACING WITH ARBITARY VALUE

```python
titanic_df['Ticket'].unique()
```

```python
titanic_df['Ticket']=titanic_df['Ticket'].fillna('C')
titanic_df['Ticket'].isnull().sum()
```

```python
titanic_df
```

# REPLACING WITH MEAN

```python
mean=titanic_df['Age'].mean()
print (mean)
titanic_df['Age']=titanic_df['Age'].fillna(mean)
titanic_df['Age']
```

# REPLACING WITH MODE

```python
titanic_df=pd.read_csv('C:/Users/GPT BANTWAL/Documents/titanic.csv')
mode=titanic_df['Ticket'].mode()[0]
print(mode)
titanic_df['Ticket']=titanic_df['Ticket'].fillna(mode)
```

```python
titanic_df['Ticket']
```

# REPLACING WITH MEDEAN


```python
titanic_df['Age']=titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Age']
```

# FORWARD AND BACKWARD FILLING OF MISSING VALUES
