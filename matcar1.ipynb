```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/GPT BANTWAL/Downloads/mtcars .csv')
df
```

```python
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color='green',rwidth=0.8)
plt.xlabel('mpg')
plt.ylabel('no of frequency')
plt.title('my histogram')
plt.show()
```

```python
plt.scatter(df['wt'],df['mpg'],color='black',marker='*')
plt.xlabel('weight of the car')
plt.ylabel('mpg')
plt.title('scatter plot')
plt.show()
```

```python
temp=df['am'].value_counts()
print(temp)
temp.plot.bar()
plt.title('ferquency distribution of transmission types')
plt.xlabel('transmission types')
plt.ylabel('count')
```

```python
plt.boxplot(df['mpg'])
plt.ylabel('mpg')
plt.title('boxplot')
plt.show()
```
